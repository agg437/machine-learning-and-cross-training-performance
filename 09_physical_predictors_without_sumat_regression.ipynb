### Physical predictors <a class="anchor" id="2"></a>

* [Returning to the index of contents](#index)

```r
handle_physical <- read.csv('handle_physical_transformed_for_multiple_linear_regression.csv')
head(handle_physical)
names(handle_physical)
```

To avoid interference with the predictors related to FFM1 and similars, we eliminated the predictors that in their name contain the FM substring.

```r
handle_physical <- handle_physical[!grepl('FM',names(handle_physical))] 
```

We delete Sumat:

```r
handle_physical <- handle_physical[!grepl('Sumat',names(handle_physical))] 
```

```r
head(handle_physical)
```

### Target variables <a class="anchor" id="3"></a>

* [Returning to the index of contents](#index)

```r
handle_target <- read.csv('handle_target_fusion_31_03_20.csv', row.name = 1)

head(handle_target)
```

Seleccionamos la variable objetivo W1-2MedDifWRelFFMI:

```r
target <- handle_target['W1.2MedDifWRelFFM1']
head(target)
```

We create the dataframe for the analysis:

```r
data_physical <- cbind(target, handle_physical)
```

## Train/test partition <a class="anchor" id ="4"></a>

* [Returning to the index of contents](#index)

```r
library(caret)
```

```r
set.seed(10)
id_df <- createDataPartition(data_physical$W1.2MedDifWRelFFM1,p = 0.85,list = FALSE)
data_physical_train <- data_physical[id_df,]
data_physical_test <- data_physical[-id_df,]

dim(data_physical_train)

dim(data_physical_test)
```

## Multiple Linear Regression <a class="anchor" id ="5"></a>

* [Returning to the index of contents](#index)

Training:

```r
model_physical_train <- lm(W1.2MedDifWRelFFM1~., data_physical_train)
```

```r
summary(model_physical_train)
```

```r
rmse <- function(obs, est) {
  sqrt(mean((obs - est)^2))
}#with this function we will compute the prediction error in train and test
```

Train error:

```r
rmse(data_physical_train[,1], model_physical_train$fitted.values)
```

Test error:

```r
pred_test <- predict(model_physical_train, newdata = data_physical_test)

rmse(data_physical_test[,1], pred_test)
```
